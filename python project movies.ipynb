```python
# import libraries 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'/Users/fairoojchowdhury/Desktop/Data alalysis/Python/movies.csv')
```

```python
# Looking at data
df.head()
```

```python
# check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
# data types for columns

df.dtypes
```

```python
# change the column data for budget, votes and gross to not include .0 at end

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',10)
```

```python
# drop duplicate data

df.drop_duplicates().head()
```

```python
# Scatter plot with Budget vs Gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('budget for film')

plt.show()
```

```python
#plot the budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#start looking at correlation

df.corr(method='pearson')
```

```python
# high correlation between Budget and Gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix')

plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```

```python
# convert company to numeric fields

df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised.head()
```

```python
df.head
```

```python
correlation_matrix = df_numerised.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix')

plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```
